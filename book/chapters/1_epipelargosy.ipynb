## shallow profiler


The OOI shallow profiler consists of two anchors on the sea floor attached by cables up to a
base *platform* as shown in the lower part of the image below. This platform is
positively buoyant, hence it *'floats'* at a depth of 200 meters below the ocean surface. 


Periodically cable is spooled out by means of a cable winch allowing an instrument-bearing 
science pod (upper right in the image) to gradually ascends from the platform depth to a 
ten-or-so meters depth below the surface. As it ascends and subsequently descencds the 
science pod acquires sensor. Once the science pod is drawn back down to the platform it 
goes into a pause mode for about an hour before the profiling process is repeated. In this 
manner the shallow profiler observes the state of the upper water column nine times per day
under good conditions, through a depth range of 10 to 200 meters.

### *Epipelagic ocean* defined


Pelagic refers to the ocean water column, surface to sea floor, usually some distance 
from the shore. *Epipelagic* is then the *upper* water column coinciding with the *sun 
illuminated* or *photic* part of the ocean.  The most common terms for the upper ocean 
are '*epipelagic zone*' and/or '*photic zone*'. 


The upper 200 meters of the water column is subject to downwelling sunlight. Below that 
depth very little light penetrates even at noon on a clear day. Sunlight is the energy 
source for primary production, particularly photosynthesis by phytoplankton. Hence *epipelagic* 
and *photic* zones are roughly synonymous terms for the marine ecosystem powered by sunlight.

## Defining an observational program


### Stability


- *stability* of the dynamic, liquid-water ocean could be amenable as a measurable phenomena
    - start with Abernathy
- stability details: dimensions in space and time
    - depth: A measurable dimension; of interest on scales of centimeters to tens of meters to kilometers
    - lateral extent (remote sensing, distributed sensors such as ARGO): meters to kilometers+
    - time: minutes to days to seasonal to annual to multi-year to climate time scale
- observables on these spatiotemporal scales
    - physical characteristics: temperature, density of water, available light, current, turbulence, opacity
    - chemistry: salinity, dissolved oxygen, concentration of inorganic carbon
    - biology: nutrient concentration particularly nitrates, particulate distribution, protein fluorescence
    - ocean-atmosphere boundary: wind, wave action, episodic events (storms); meters to hundreds of kilometers
    - large scale perturbation: sea state, upwelling, eddies, terrigenous influence (runoff)
- inferred structure
    - profiles, stratification
        - mixed layer depth, clines
        - asymptotes in the lower epipelagic
        - inversion layers, lensing


These observables can often be related to stability through deviation with time. The narrative 
consequently moves towards two ideas: What is observed? And how frequently?

### Thermohaline circulation

Surface currents in the ocean are driven by prevailing winds. 
The ocean below the epipelagic, in contrast, is increasingly indifferent to the ocean-atmosphere
interface: Wind has less influence. Rather, circulation in the deep ocean is driven by gradients 
in temperature and salinity ('*thermohaline*'), the two important factors determining water density. 
In this regime the important time scale is a range from centuries to millenia. 
The observational record for the regional cabled array, in contrast, is approaching one decade.

### Coincidence and persistence


The figure below shows temperature versus depth in the epipelagic: from 200 meters 
to near surface. 

````python
Image(filename='./../img/sp_ascent_vs_descent_temperature.png', width=600)
````

These data were acquired by a shallow profiler in ascent and descent phases
(red and green respectively). The low temperature excursion feature seen from 65 to 
85 meters is *persistent* between the two observation phases. These were separated 
in time by about zero minutes at the top of the profile and by over an hour at
the bottom of the profile. Note that from 100 meters downward the profile 
structures are similar but exhibit a relative offset.


- *Persistence*: Ocean structure that persists in time
    - for example across multiple consecutive observations as shown above.
- *Coincidence*: Data structure apparent across multiple sensor streams
    - for example a feature or 'jag' apparent in both temperature and salinity data

### Comparison with remote sensing


Suppose a near-surface temperature anomaly persists in shallow profiler data for 50 
hours with a stable coincident current measurement of 2 km / hour south. This suggests
structure on a lateral scale of 100 km which might be amenable to confirmation by
means of satellite data.


- The ocean water column is understood to be stratified
    - The upper layer is the *mixed layer* (often homogeneous, 50-100m)
    - The transitional layer below the mixed layer is the *pycnocline*
    - The *lower epipelagic*: Photic zone from the lower pycnocline downward
